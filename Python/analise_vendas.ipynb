```python
def preencher_uf(row):
    if pd.isna(row['UF']) and row['CodigoCidade'] == 20027:
        return 'RS'
    elif pd.isna(row['UF']) and row['CodigoCidade'] == 15428:
        return 'SP'
    elif pd.isna(row['UF']) and row['CodigoCidade'] == 33005:
        return 'RJ'
    else:
        return row['UF']
    
master_stores['UF'] = master_stores.apply(preencher_uf, axis=1)
master_stores
```

```python
# Primeiro, juntamos os dados de vendas com informações dos clientes para ter a classe social
sales_with_social_class = pd.merge(raw_data, master_clients[['CustomerID', 'ClasseSocial']], on='CustomerID', how='left')

# Agora, calculamos o valor total de vendas por classe social
sales_by_social_class = sales_with_social_class.groupby('ClasseSocial')['ValueSold'].sum().reset_index()

# Encontramos a classe social com maior valor de vendas
top_social_class = sales_by_social_class.sort_values('ValueSold', ascending=False).iloc[0]['ClasseSocial']

# Filtramos as vendas para apenas clientes da classe social que mais comprou
top_class_sales = sales_with_social_class[sales_with_social_class['ClasseSocial'] == top_social_class]

# Agora, precisamos contar os clientes que compraram peças em promoção em cada UF
# Primeiro, juntamos os dados de vendas com informações das lojas para ter a UF
top_class_sales_with_uf = pd.merge(top_class_sales, master_stores[['StoreID', 'UF']], on='StoreID', how='left')

# Filtramos para apenas vendas em promoção
promo_sales = top_class_sales_with_uf[top_class_sales_with_uf['Promo'] == 1]

# Contamos os clientes únicos que compraram em promoção em cada UF
promo_sales_by_uf = promo_sales.groupby('UF')['CustomerID'].nunique().reset_index()

promo_sales_by_uf.rename(columns={'CustomerID': 'UniqueCustomersBoughtOnPromo'}, inplace=True)

top_social_class, promo_sales_by_uf
```

```python
# Filtrando as transações para peças da cor azul
blue_sales = raw_data[raw_data['Cor'] == 'Azul']

# Juntando os dados de vendas com informações das lojas para obter a UF
blue_sales_with_uf = pd.merge(blue_sales, master_stores[['StoreID', 'UF']], on='StoreID', how='left')

# Encontrando clientes que compraram peças azuis em apenas uma UF
# Contamos quantas UFs únicas cada cliente comprou
unique_uf_purchases = blue_sales_with_uf.groupby('CustomerID')['UF'].nunique().reset_index()

# Filtramos para clientes que compraram em apenas uma UF
single_uf_customers = unique_uf_purchases[unique_uf_purchases['UF'] == 1]['CustomerID']

# Filtramos as vendas de peças azuis para apenas esses clientes
blue_sales_single_uf = blue_sales_with_uf[blue_sales_with_uf['CustomerID'].isin(single_uf_customers)]

# Agora, determinamos qual tamanho vendeu mais peças azuis, contando as unidades vendidas
top_size_blue_sales = blue_sales_single_uf.groupby('Tamanho')['UnitsSold'].sum().reset_index()

# Encontramos o tamanho com maior número de unidades vendidas
top_selling_size = top_size_blue_sales.sort_values('UnitsSold', ascending=False).iloc[0]['Tamanho']

top_selling_size, top_size_blue_sales
```

```python

# Filtrando as vendas para peças da cor vermelha e adicionando o ano da venda
red_sales = raw_data[raw_data['Cor'] == 'Vermelha'].copy()
red_sales['Year'] = red_sales['Date'].dt.year

# Agregando as vendas por ano, tipo de item (Descrição) e UF
red_sales_by_year = red_sales.groupby(['Year', 'Descrição', 'StoreID'])['UnitsSold'].sum().reset_index()

# Juntando os dados de vendas com informações das lojas para obter a UF
red_sales_by_year_with_uf = pd.merge(red_sales_by_year, master_stores[['StoreID', 'UF']], on='StoreID', how='left')

# Separando as vendas por ano
red_sales_2022 = red_sales_by_year_with_uf[red_sales_by_year_with_uf['Year'] == 2022]
red_sales_2023 = red_sales_by_year_with_uf[red_sales_by_year_with_uf['Year'] == 2023]

# Agregando as vendas por descrição e UF para cada ano
total_red_sales_2022 = red_sales_2022.groupby(['Descrição', 'UF'])['UnitsSold'].sum().reset_index()
total_red_sales_2023 = red_sales_2023.groupby(['Descrição', 'UF'])['UnitsSold'].sum().reset_index()

# Juntando as vendas de 2022 e 2023 para comparar
comparison_sales = pd.merge(total_red_sales_2022, total_red_sales_2023, on=['Descrição', 'UF'], how='inner', suffixes=('_2022', '_2023'))

# Filtrando para itens que venderam mais em 2023 do que em 2022 em todas as UFs
comparison_sales['MoreIn2023'] = comparison_sales['UnitsSold_2023'] > comparison_sales['UnitsSold_2022']
items_sold_more_in_2023 = comparison_sales.groupby('Descrição')['MoreIn2023'].all().reset_index()

# Encontrando os itens que satisfazem a condição
top_items = items_sold_more_in_2023[items_sold_more_in_2023['MoreIn2023'] == True]['Descrição']

# Se houver mais de um item, escolheremos um para continuar a análise
if not top_items.empty:
    selected_item = top_items.iloc[0]

    # Filtrando as vendas de 2022 para o RS e para o item selecionado
    sales_in_rs_2022 = red_sales_2022[(red_sales_2022['Descrição'] == selected_item) & (red_sales_2022['UF'] == 'RS')]

    # Calculando o valor total vendido para esse item no RS em 2022
    total_value_sold_in_rs_2022 = sales_in_rs_2022['UnitsSold'].sum()
else:
    selected_item = None
    total_value_sold_in_rs_2022 = 0

selected_item, total_value_sold_in_rs_2022
```
